```python
def city_cost_category(x):
    if len(x) < 4:
        return pd.Series(["Medium"] * len(x), index=x.index)

    return pd.qcut(
        x.rank(method="first"),
        q=4,
        labels=["Budget", "Medium", "Premium", "Luxury"]
    ).astype(str)

data["City wise Cost Category"] = (
    data.groupby("City")["Average Cost INR"]
    .transform(city_cost_category)
)
```

### Group nearby restaurants

```python
import math
import numpy as np
from sklearn.cluster import KMeans

def get_cluster_count_by_area(group, target_area_km2=25, max_clusters=30):
    n_restaurants = len(group)

    unique_locations = group[["Latitude", "Longitude"]].drop_duplicates()
    n_unique_locations = len(unique_locations)

    if n_restaurants < 2 or n_unique_locations < 2:
        return 1

    lat = group["Latitude"].astype(float)
    lon = group["Longitude"].astype(float)

    lat_min, lat_max = lat.quantile([0.05, 0.95])
    lon_min, lon_max = lon.quantile([0.05, 0.95])

    height_km = (lat_max - lat_min) * 111
    width_km = (lon_max - lon_min) * 111 * np.cos(np.deg2rad(lat.mean()))

    city_area_km2 = abs(height_km * width_km)

    n_clusters = math.ceil(city_area_km2 / target_area_km2)

    return min(
        max(n_clusters, 1),
        max_clusters,
        n_restaurants,
        n_unique_locations
    )


data["Location Cluster"] = 0

valid_location = data[["Latitude", "Longitude"]].notna().all(axis=1)

for city, group in data[valid_location].groupby("City"):
    n_clusters = get_cluster_count_by_area(group)

    if n_clusters == 1:
        data.loc[group.index, "Location Cluster"] = 0
    else:
        kmeans = KMeans(
            n_clusters=n_clusters,
            random_state=42,
            n_init=10
        )

        data.loc[group.index, "Location Cluster"] = kmeans.fit_predict(
            group[["Latitude", "Longitude"]]
        )

data["City Location Cluster"] = (
    data["City"].astype(str) + "_" + data["Location Cluster"].astype(str)
)
```
